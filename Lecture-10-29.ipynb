### Auditory scales
- Linear
- Logarithmic
- Mel Scale
- Bark Scale
- ERB

```python
### plots of different auditory scales
a = np.linspace(0,10000, 100, endpoint=True)
b_linear = a
b_log = np.log2(a[1:])
b_mel = 2595 * np.log10(1+a/100)
b_bark = 13*np.arctan(0.00076*a)+3.5*np.arctan((a/7500)**2)
b_erb = 21.4*np.log10(0.00437*a +1)
fig, ax = plt.subplots(3,2, figsize=(10,8))

ax[0,0].plot(a,b_linear)
ax[1,0].plot(a[1:], b_log)
ax[2,0].plot(a, b_mel)
ax[0,1].plot(a, b_bark)
ax[1,1].plot(a, b_erb)
ax[2,1].plot(a[1:], b_log)
#ax[2,1].plot(a, b_mel)
ax[2,1].plot(a, b_bark)
ax[2,1].plot(a, b_erb)
ax[2,1].legend(["log", "bark", "erb"])
ax[0,0].set_xlabel("Frequency/Hz(Physical)")
ax[0,0].set_ylabel("Frequency/Hz(Auditory)")
ax[0,0].set_title("Linear")

ax[1,0].set_xlabel("Frequency/Hz(Physical)")
ax[1,0].set_ylabel("Frequency/Hz(Auditory)")
ax[1,0].set_title("Logarithmic")

ax[2,0].set_xlabel("Frequency/Hz(Physical)")
ax[2,0].set_ylabel("Frequency/Hz(Auditory)")
ax[2,0].set_title("Mel Scale")

ax[0,1].set_xlabel("Frequency/Hz(Physical)")
ax[0,1].set_ylabel("Frequency/Hz(Auditory)")
ax[0,1].set_title("Bark Scale")

ax[1,1].set_xlabel("Frequency/Hz(Physical)")
ax[1,1].set_ylabel("Frequency/Hz(Auditory)")
ax[1,1].set_title("ERB Scale")

ax[2,1].set_xlabel("Frequency/Hz(Physical)")
ax[2,1].set_ylabel("Frequency/Hz(Auditory)")
ax[2,1].set_title("LOG/BARK/ERB Scale")

plt.tight_layout()
```
